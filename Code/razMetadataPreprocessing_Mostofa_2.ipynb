```python
# import sys
# filename_arg='/content/A/A_ThisIsMyBear.docx'
# print('this is the fihlesname' + str(filename_arg))

def get_book_text(filename):
#   file_ref = open(filename,"rb")
  doc = docx.Document(filename)

  text_lines = []
  text_lines_2 = []
  bold_lines = []
  lines_needed = []
  lines_needed_2 = []
  lines_needed_3 = []
  abc = []
  c = []

  regex = re.compile('[.?]') 

  for paragraph in doc.paragraphs:
    for run in paragraph.runs:
      text_lines.append(run.text)

  for paragraph in doc.paragraphs:
    for run in paragraph.runs:
      text_lines_2.append(run.text)
    
    
  
  index = [idx for idx, s in enumerate(text_lines_2) if '.mp3' in s][0]
  title_extract_1 = text_lines_2[index+1]
  title_extract_2 = text_lines_2[index+2]

  if(len(title_extract_1) == 0): 
    title_extract_1 = title_extract_2.split(' ')[0] 
  #print(lines_needed) 
  
  r1 = re.compile(' ') 

  if(r1.search(title_extract_1) == None):
    pass      
  else: 
    title_extract_1 = title_extract_1.split(' ')[0] 
    
    
    
    
  index = [idx for idx, s in enumerate(text_lines) if '.mp3' in s][1]
  del text_lines[0:index]

  for paragraph in doc.paragraphs:
    for run in paragraph.runs:
      if run.bold: 
        bold_lines.append(run.text)
  
  for line in text_lines:
    if line not in bold_lines:
      lines_needed.append(line)
    
  lines_needed = [x for x in lines_needed if x]
  
#   index = [idx for idx, s in enumerate(text_lines_2) if '.mp3' in s][0]
#   title_extract_1 = text_lines_2[index+1]
#   title_extract_2 = text_lines_2[index+2]

#   if(len(title_extract_1) == 0): 
#     title_extract_1 = title_extract_2.split(' ')[0] 
#   #print(lines_needed) 
  
#   r1 = re.compile(' ') 

#   if(r1.search(title_extract_1) == None):
#     pass      
#   else: 
#     title_extract_1 = title_extract_1.split(' ')[0] 
    
    
  for line in lines_needed:
    if(regex.search(line) == None):
        pass
#         print("String is accepted") 
          
    else: 
#         print("String is not accepted.") 
        lines_needed_2.append(line)
  
  special_characters = "_"
    
  for line_2 in lines_needed_2:
    if any(c in special_characters for c in line_2):
        abc = re.split('(?<=[.!?]) +',line_2)
        for line_1 in abc:
            if any(c in special_characters for c in line_1):
                pass
            else: 
                lines_needed_3.append(line_1)
                lines_needed_3.append("<start>")

    else: 
        lines_needed_3.append(line_2)
        lines_needed_3.append("<start>")


    
#   lines_needed_2 = [xy1 for xy1 in lines_needed_2 if xy1]

#   special_characters = "_"
#   r2 = re.compile() 

#   for line_2 in lines_needed_2:
#     if any(c in special_characters for c in line_2):
#         pass
# #         abc = r2.split('(?<=[.!?]) +',line_2)
# #       for line_1 in abc:
# #         if any(c in special_characters for c in line_1):
# #           pass
# #         else: 
# #           lines_needed_3.append(line_1)

#     else: 
#       lines_needed_3.append(line_2)
      
    
  return [lines_needed_3, bold_lines,title_extract_1]  
#   return lines_needed_2 
```
